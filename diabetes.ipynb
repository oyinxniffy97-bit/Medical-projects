## Attempt 2: SMOTE Oversampling

SMOTE was applied to the training set only (test set left untouched to preserve 
real-world class proportions), generating synthetic examples until all three classes 
were equally represented (170,962 each).

Results improved meaningfully for class 2 (diabetes): recall rose from 0.20 to 0.49, 
at the cost of overall accuracy dropping from 0.84 to 0.78 — an acceptable tradeoff, 
since correctly identifying diabetes cases matters more than raw accuracy in a 
screening context.

Class 1 (prediabetes) recall remained near 0 (0.01) even with balanced training data, 
suggesting this class may lack a clear separable signal within the available features, 
rather than simply needing more examples.

```python
from sklearn.linear_model import LogisticRegression

model_lr = LogisticRegression(max_iter=1000, random_state=42)
model_lr.fit(X_train_resampled, y_train_resampled)

y_pred_lr = model_lr.predict(X_test)
print(classification_report(y_test, y_pred_lr))
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_resampled)
X_test_scaled = scaler.transform(X_test)

model_lr.fit(X_train_scaled, y_train_resampled)
y_pred_lr = model_lr.predict(X_test_scaled)
print(classification_report(y_test, y_pred_lr))
```

## Reframing as Binary Classification

Across three different algorithms (Random Forest, Logistic Regression, XGBoost), 
class 1 (prediabetes) consistently failed to be identified, with recall never 
exceeding 0.19. Since this pattern held across structurally different models, 
it points to a limitation in feature signal rather than model choice.

Given that prediabetes and diabetes both warrant similar clinical follow-up 
(lifestyle intervention, monitoring, further testing), the target was reframed 
as binary: 0 = no diabetes, 1 = at risk (prediabetes or diabetes combined). 
This better reflects a practical screening use case and removes the 
unresolvable three-class signal problem.

```python
y_binary = df['Diabetes_012'].astype(int).replace({1: 1, 2: 1})  # 0 stays 0, 1&2 become 1
# equivalently: y_binary = (df['Diabetes_012'].astype(int) > 0).astype(int)

print(y_binary.value_counts(normalize=True))
```

```python
from sklearn.model_selection import train_test_split

X_train_b, X_test_b, y_train_b, y_test_b = train_test_split(
    X, y_binary, test_size=0.2, random_state=42, stratify=y_binary
)

ordinal_cols = ['GenHlth', 'Education', 'Income', 'Age']
X_train_b[ordinal_cols] = X_train_b[ordinal_cols].astype(int)
X_test_b[ordinal_cols] = X_test_b[ordinal_cols].astype(int)
```

```python
from imblearn.over_sampling import SMOTE

smote_b = SMOTE(random_state=42)
X_train_b_resampled, y_train_b_resampled = smote_b.fit_resample(X_train_b, y_train_b)

print(y_train_b_resampled.value_counts())
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report

model_xgb_b = XGBClassifier(random_state=42, eval_metric='logloss')
model_xgb_b.fit(X_train_b_resampled, y_train_b_resampled)

y_pred_xgb_b = model_xgb_b.predict(X_test_b)
print(classification_report(y_test_b, y_pred_xgb_b))
```

## Binary Model Results

SMOTE was reapplied to the binary training set to correct for the remaining 
imbalance (no diabetes still outnumbers at-risk cases).

| Model | Accuracy | Recall (at risk) | Precision (at risk) | F1 (at risk) | Macro F1 |
|---|---|---|---|---|---|
| XGBoost + SMOTE | 0.77 | 0.56 | 0.36 | 0.43 | 0.65 |

Reframing the problem as binary substantially improved model performance for 
detecting at-risk individuals. Recall for the at-risk class rose to 0.56, 
compared to a best result of 0.49-0.54 across three-class attempts, and macro 
F1 improved from 0.43 to 0.65 — confirming that the three-class formulation's 
core limitation was an unrecoverable middle category (prediabetes) rather than 
model choice. This binary framing also better matches a practical screening 
use case, where distinguishing "no risk" from "needs follow-up" is more 
actionable than separating two categories that call for similar clinical response.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

model_rf_b = RandomForestClassifier(random_state=42)
model_rf_b.fit(X_train_b_resampled, y_train_b_resampled)

y_pred_rf_b = model_rf_b.predict(X_test_b)
print(classification_report(y_test_b, y_pred_rf_b))
```

## Model Comparison — Binary Classification

| Model | Accuracy | Recall (at risk) | Precision (at risk) | F1 (at risk) | Macro F1 |
|---|---|---|---|---|---|
| XGBoost + SMOTE | 0.77 | 0.56 | 0.36 | 0.43 | 0.65 |
| Random Forest + SMOTE | 0.79 | 0.46 | 0.37 | 0.41 | 0.64 |

Both models perform similarly overall, but XGBoost recalls substantially more 
at-risk cases (0.56 vs 0.46) at a small cost to overall accuracy (0.77 vs 0.79). 
In a health screening context, missing a true at-risk individual carries a 
higher cost than a false positive, since a false positive simply results in 
unnecessary follow-up testing while a false negative allows a real health risk 
to go undetected. **XGBoost + SMOTE was selected as the final model** on this basis.

## Conclusion

This analysis progressed through several stages: cleaning and correctly typing 
the BRFSS 2015 dataset, establishing that naive accuracy was misleading given 
severe class imbalance (84% "no diabetes"), and testing multiple approaches to 
address it. Reweighting (class_weight='balanced') produced negligible improvement, 
while SMOTE oversampling meaningfully improved detection of diabetes cases across 
all three algorithms tested (Random Forest, Logistic Regression, XGBoost) — 
though the middle category, prediabetes, remained undetectable regardless of 
algorithm or resampling strategy, indicating a feature signal limitation rather 
than a modeling one.

Reframing the problem as binary classification (no diabetes vs. at risk) resolved 
this limitation and produced the strongest overall results, with the final 
XGBoost + SMOTE model achieving 0.56 recall and 0.65 macro F1 for identifying 
individuals at risk of diabetes or prediabetes — a practically useful result for 
a screening context, where the priority is minimizing missed at-risk cases 
rather than maximizing raw accuracy.

## Limitations and Future Work

- #Feature set is limited to BRFSS survey responses; additional clinical data 
  (e.g. fasting glucose, HbA1c) would likely improve separability, particularly 
  for the prediabetes category.
- Precision for the at-risk class remains modest (0.36), meaning a meaningful 
  number of false positives would occur in practice; threshold tuning or 
  cost-sensitive evaluation could help balance this against recall.
- Feature importance analysis was not performed in this notebook but would be 
  a natural next step to understand which health indicators drive the model's 
  predictions.

```python
import matplotlib.pyplot as plt
import numpy as np

models = ['Baseline RF', 'RF+weight', 'RF+SMOTE', 'LogReg+SMOTE', 'XGB+SMOTE', 
          'Binary XGB', 'Binary RF']
accuracy = [0.84, 0.84, 0.78, 0.65, 0.74, 0.77, 0.79]
macro_f1 = [0.40, 0.38, 0.43, 0.42, 0.43, 0.65, 0.64]

x = np.arange(len(models))
width = 0.35

fig, ax = plt.subplots(figsize=(11, 5))
ax.bar(x - width/2, accuracy, width, label='Accuracy', color='#4C72B0')
ax.bar(x + width/2, macro_f1, width, label='Macro F1', color='#DD8452')

ax.set_ylabel('Score')
ax.set_title('Model Comparison: Accuracy vs Macro F1')
ax.set_xticks(x)
ax.set_xticklabels(models, rotation=30, ha='right')
ax.legend()
ax.axhline(y=0.5, color='gray', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(9, 5))

recall_data = {
    'Baseline RF':    [0.97, 0.00, 0.20],
    'RF+weight':      [0.97, 0.00, 0.16],
    'RF+SMOTE':       [0.84, 0.01, 0.49],
    'LogReg+SMOTE':   [0.66, 0.19, 0.60],
    'XGB+SMOTE':      [0.79, 0.06, 0.54],
}

classes = ['No Diabetes (0)', 'Prediabetes (1)', 'Diabetes (2)']
x = np.arange(len(classes))
width = 0.15

for i, (model, recalls) in enumerate(recall_data.items()):
    ax.bar(x + i*width, recalls, width, label=model)

ax.set_ylabel('Recall')
ax.set_title('Recall by Class Across Models — Prediabetes Consistently Undetected')
ax.set_xticks(x + width*2)
ax.set_xticklabels(classes)
ax.legend(fontsize=8)
plt.tight_layout()
plt.show()
```
